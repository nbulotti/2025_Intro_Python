```python
# Replicate this figure
# Temperature values are originally in Kelvin units -> Convert to degree

fig,(ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8,4), gridspec_kw={'width_ratios':[5,1.5]})
# create figure with 1 row, 2 columns, set subplot width ratios and figure size
ctemp= ax1.contourf(lon, lat, temp-273.15, cmap='magma', levels=np.linspace(-30,40,15), extend='both')
# filled contour plot of temperature in Celsius, specify colormap, contour levels, extend both ends
ax1.contour(lon, lat, temp-273.15, colors='w', levels=[-10], extend='both')
# add contour line at -10¬∞C in white, extend both ends
ax1.set_xlabel('Longitude')  # set x-axis label
ax1.set_ylabel('Latitude')   # set y-axis label
ax1.set_title('Current Global Temperature')  # set subplot title
fig.colorbar(ctemp, ax=ax1, label='$^o$C')
# add colorbar for contour plot, link to ax1, label in ¬∞C
ax2.plot(np.mean(temp-273.15, axis=1), lat, lw=2, c='k')
# plot zonal (longitudinal) mean temperature, line width 2, black color
ax2.set_xlabel(r'$^{o}$C')  # set x-axis label for zonal mean
ax2.set_ylabel('Latitude')   # set y-axis label
ax2.set_title('Zonal Mean Temperature')  # set subplot title
plt.tight_layout()  # automatically adjust subplot spacing
plt.show()          # display the figure
```

### Part II: Scatter Plots to Visualize Earthquake Data
Here, we will make a map plot of earthquakes from a USGS catalog of historic large earthquakes. Color the earthquakes by `log10(depth)` and adjust the marker size to be `magnitude`

```python
fname = pooch.retrieve(
    "https://unils-my.sharepoint.com/:u:/g/personal/tom_beucler_unil_ch/EW1bnM3elHpAtjb1KtiEw0wB9Pl5w_FwrCvVRlnilXHCtg?download=1",
    known_hash='22b9f7045bf90fb99e14b95b24c81da3c52a0b4c79acf95d72fbe3a257001dbb',
    processor=pooch.Unzip()
)[0]

earthquakes = np.genfromtxt(fname, delimiter='\t')
depth = earthquakes[:, 8]
magnitude = earthquakes[:, 9]
latitude = earthquakes[:, 20]
longitude = earthquakes[:, 21]
```

Hint: Check out the [Scatter Plots subsection](#Scatter) and consider reading the documentation for [`plt.scatter`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) and [`plt.colorbar`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.colorbar.html).

```python
# This is the function to the set the ticks in the colorbar to be in scientific notations.
# You can directly use this function and don't need to change anything in this cell.
import matplotlib.ticker as ticker
def fmt(x, pos):
    a, b = '{:.2e}'.format(x).split('e')
    b = int(b)
    return r'$10^{{{}}}$'.format(a)
```

```python
# Replicate the figure here

fig, ax = plt.subplots(figsize=(8,4))
# create figure and axes with specified size
ctemp = ax.scatter(longitude, latitude, s=magnitude, c=np.log10(depth), marker='o', cmap='viridis', vmin=0, vmax=2.5)
# scatter plot: lon vs lat, marker size proportional to magnitude, color by log10(depth), colormap viridis, set min/max color
ax.set_xlabel('Longitude')  # set x-axis label
ax.set_ylabel('Latitude')   # set y-axis label
plt.grid()                  # add gridlines
fig.colorbar(ctemp, ax=ax, label="Depth (Km)", format=ticker.FuncFormatter(fmt))
# add colorbar for scatter, linked to ax, with label and custom formatter
ax.set_title('Location of Significant Earthquake With Size Representing Relative Magnitude')
plt.show()                  # display the figure
```

```python
import matplotlib.pyplot as plt  # import plotting library
#import cartopy.feature as cfeature
import numpy as np  # import numerical library

######################################################################################################################################################
# Replicate the figure here
fig,ax = plt.subplots(1,2,figsize=(18, 12),subplot_kw={'projection': ccrs.SouthPolarStereo()})  # create figure with 2 subplots using South Polar Stereo projection
# The rest doesn't change:
ax[0].set_extent([-180, 180, -90, -55], ccrs.PlateCarree())  # set map limits in geographic coordinates

######################################################################################################################################################
# Add Land and gridlines
# Hint: Check edgecolor to show coast, and facecolor to color the continent
# Hint: https://matplotlib.org/stable/gallery/color/named_colors.html for the names of the colors available
######################################################################################################################################################
ax[0].add_feature(cartopy.feature.LAND, edgecolor='black',facecolor='lightgray')  # add land with black border and gray fill
ax[0].gridlines(lw=1,color='gray')  # add gridlines with specified line width and color

######################################################################################################################################################
# Add Contours
# Hint: X (Y) data in the pcolormesh should be the longitude and latitude of the sea ice data.
# Hint: Check the shape of the var_toplot1, var_toplot2 before plotting
# Hint: longitude and latitude can be accessed with seaice1.variables['longitude'][:], seaice1.variables['latitude'][:]
# Hint: Check this website: https://matplotlib.org/stable/tutorials/colors/colormaps.html for available colormaps
######################################################################################################################################################

print(var_toplot1.shape)  # check shape of first sea ice array (time, y, x)
print(var_toplot2.shape)  # check shape of second sea ice array

icext1= ax[0].pcolormesh(
    seaice1.variables['longitude'][:],  # X axis = longitude
    seaice1.variables['latitude'][:],   # Y axis = latitude
    np.ma.masked_greater(var_toplot1[0],1),  # mask values >1 (invalid sea ice concentration)
    transform=ccrs.PlateCarree(),  # input data are in PlateCarree (lat/lon)
    vmin=0,vmax=1,  # color limits
    cmap="cividis"  # colormap
)
plt.colorbar(icext1,ax=ax[0],fraction=0.046, pad=0.04,label='Sea Ice Concentratiion')  # add colorbar for first plot
ax[0].set_title('2017-08-01')  # add title

ax[1].set_extent([-180, 180, -90, -55], ccrs.PlateCarree())  # set map extent for second subplot
ax[1].add_feature(cartopy.feature.LAND, edgecolor='black',facecolor='lightgray')  # add land with edges
ax[1].gridlines(lw=1,color='gray')  # add gridlines

icext2 = ax[1].pcolormesh(
    seaice2.variables['longitude'][:],  # X axis = longitude
    seaice2.variables['latitude'][:],   # Y axis = latitude
    np.ma.masked_greater(var_toplot2[0],1),  # mask values >1
    transform=ccrs.PlateCarree(),  # input data CRS
    vmin=0,vmax=1,  # color range
    cmap="cividis"  # colormap
)
plt.colorbar(icext2,ax=ax[1],label='Sea Ice Concentration',fraction=0.046, pad=0.04)  # add colorbar
ax[1].set_title('2017-12-31')  # set title for second subplot
plt.show()  # render figure
```

```python
import pandas as pd
fname = pooch.retrieve(
    "https://unils-my.sharepoint.com/:x:/g/personal/tom_beucler_unil_ch/Ea9h1j2_wYpEtuX5waZZWpsBt2zh3lvGUBFisvA8dFG5Eg?download=1",
    known_hash='84d455fb96dc8f782fba4b5fbe56cb8970cab678f07c766fcba1b1c4674de1b1')
usgs_2014 = pd.read_csv(fname)
```

```python
######################################################################################################################################################
# Run these as is. We are removing missing values in the data and filter data points with stronger earthquakes (magnitude > 4)
usgs_2014_nonan = usgs_2014.dropna()
usgs_2014_large = usgs_2014_nonan[usgs_2014_nonan['mag']>4]
```

```python
fig,ax = plt.subplots(1,1,figsize=(12, 12),subplot_kw={'projection': ccrs.Robinson()})
# The rest doesn't change:
ax.set_extent([-140, -60, 12,70], ccrs.PlateCarree())

######################################################################################################################################################
# Add Land, Ocean, Rivers, and State Lines
# Hint: Land linewidth = 3; States: edgecolor='gray',linewidth=1.25; lakes/ocean/rivers: no change
# Hint: https://matplotlib.org/stable/gallery/color/named_colors.html for the names of the colors available
######################################################################################################################################################
ax.add_feature(cartopy.feature.LAND, edgecolor='black',linewidths=3)
ax.add_feature(cartopy.feature.OCEAN)
ax.add_feature(cartopy.feature.RIVERS)
ax.add_feature(cartopy.feature.STATES)
ax.add_feature(cartopy.feature.COASTLINE,edgecolor='gray',linewidth=1.25)
ax.gridlines(lw=1,color='white')

######################################################################################################################################################
# Add Scatter points for earthquake data
# Hint: usgs_2014_large['longitude'] to access logitude data, usgs_2014_large['latitude'] to access latitude data, usgs_2014_large['mag'] to access Earthquake magnitudes
# Hint: https://matplotlib.org/stable/gallery/color/named_colors.html for the names of the colors available
######################################################################################################################################################
equake = ax.scatter(usgs_2014_large['longitude'],usgs_2014_large['latitude'],s=62,c=usgs_2014_large['mag'],marker='o',cmap='Reds',edgecolors='black',linewidths=2, transform=ccrs.PlateCarree())
plt.colorbar(equake,ax=ax,fraction=0.016, pad=0.04,label='Earthquake Magnitude')
ax.set_title('Eaerthquake > 4')
plt.show()
```
